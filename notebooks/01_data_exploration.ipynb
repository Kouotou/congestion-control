## 1. Setup and load data
Make sure to run this notebook from the repository root (`congestioncontrol/`) so relative paths resolve correctly.

```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure repo root is on path when notebook runs from notebooks/
repo_root = Path('..').resolve()
sys.path.insert(0, str(repo_root))

data_path = repo_root / 'data' / 'weblog.csv'
# If no dataset exists, create a small synthetic sample for demo purposes
if not data_path.exists():
    df = pd.DataFrame({
        'timestamp': pd.date_range(start='2026-01-01', periods=120, freq='5min'),
        'concurrent_users': np.random.randint(50, 300, size=120),
        'requests_per_second': np.random.uniform(5, 50, size=120),
        'traffic_volume_mb': np.random.uniform(20, 500, size=120),
        'network_utilization': np.random.uniform(20, 95, size=120),
    })
    (repo_root / 'data').mkdir(parents=True, exist_ok=True)
    df.to_csv(data_path, index=False)
else:
    df = pd.read_csv(data_path, parse_dates=['timestamp'])

print('Loaded data from', data_path)
print('Shape:', df.shape)
df.head()
```

```python
# Basic info and statistics
print(df.info())
display(df.describe())
```

## 2. Quick visualizations
Plot a few KPIs to inspect trends and seasonality. Add images to the presentation by saving PNGs from cells.

```python
plt.figure(figsize=(10,4))
plt.plot(df['timestamp'], df['concurrent_users'], label='concurrent_users')
plt.xlabel('time')
plt.ylabel('users')
plt.title('Concurrent users over time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
# Save placeholder screenshot (presentation use)
(repo_root / 'frontend' / 'screenshots').mkdir(parents=True, exist_ok=True)
plt.savefig(repo_root / 'frontend' / 'screenshots' / 'figure_1_concurrent_users.png')
```

## 3. Preprocessing notes
Demonstrate resampling, missing-value handling, and feature engineering (lags, rolling means).

```python
# Example: create lag features and rolling stats
df = df.sort_values('timestamp').reset_index(drop=True)
df['users_lag_1'] = df['concurrent_users'].shift(1)
df['users_roll_3'] = df['concurrent_users'].rolling(window=3, min_periods=1).mean()
display(df.head())
```

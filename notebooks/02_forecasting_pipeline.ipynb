## Instructions
Run this notebook from repository root so Python imports from `app` resolve correctly. If running from `notebooks/`, set `sys.path` as shown.

```python
import sys
from pathlib import Path
import pandas as pd

# ensure repo root is on path when notebook runs from notebooks/
repo_root = Path('..').resolve()
sys.path.insert(0, str(repo_root))

from app.ml.pipeline import ForecastPipeline

# Example: use a small CSV (creates a sample if missing)
csv_path = repo_root / 'data' / 'weblog.csv'
if not csv_path.exists():
    # create small sample (same as in EDA)
    df = pd.DataFrame({
        'timestamp': pd.date_range(start='2026-01-01', periods=60, freq='5min'),
        'concurrent_users': range(100,160),
        'requests_per_second': [10 + i*0.2 for i in range(60)],
        'traffic_volume_mb': [40 + i*1.2 for i in range(60)],
        'network_utilization': [30 + i*0.3 for i in range(60)],
    })
    csv_path.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(csv_path, index=False)
    print('Wrote sample CSV to', csv_path)
else:
    print('Using existing CSV:', csv_path)

csv_path = str(csv_path)
pipeline = ForecastPipeline(model_dir=str(repo_root / 'backend' / 'app' / 'ml' / 'models'))

# Quick run: train models from CSV (may take time depending on selected models)
# Use a small dataset or reduce candidate models for demo speed.
result = pipeline.train_from_csv(csv_path, target_column='concurrent_users')
print('Training result summary:')
print(result)
```

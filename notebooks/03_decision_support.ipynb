# Notebook 3 — Decision support demonstrations

This notebook shows how the decision support functions (congestion scoring, QoE prediction, QoS recommendation, resource allocation, and what-if simulation) produce results from KPI inputs.

```python
import sys
from pathlib import Path
repo_root = Path('..').resolve()
sys.path.insert(0, str(repo_root))

from app.services.decision_support import (
    predict_congestion,
    predict_qoe,
    recommend_qos,
    recommend_resource_allocation,
    run_what_if_simulation,
)

# Example inputs
input_example = {
    'concurrent_users': 150,
    'requests_per_second': 35.5,
    'traffic_volume_mb': 320.0,
    'network_utilization': 75.0,
    'latency_ms': 120.0,
    'jitter_ms': 20.0,
    'packet_loss_pct': 0.6,
    'throughput_mbps': 80.0,
    'response_time_ms': 140.0,
    'timeout_rate_pct': 0.8,
    'expected_session_duration_minutes': 30,
}

print('Congestion:')
print(predict_congestion(
    concurrent_users=input_example['concurrent_users'],
    requests_per_second=input_example['requests_per_second'],
    traffic_volume_mb=input_example['traffic_volume_mb'],
    network_utilization=input_example['network_utilization'],
))

print('
QoE:')
print(predict_qoe(
    latency_ms=input_example['latency_ms'],
    jitter_ms=input_example['jitter_ms'],
    packet_loss_pct=input_example['packet_loss_pct'],
    throughput_mbps=input_example['throughput_mbps'],
    response_time_ms=input_example['response_time_ms'],
    timeout_rate_pct=input_example['timeout_rate_pct'],
))

print('
QoS recommendation:')
print(recommend_qos(
    predicted_concurrent_users=120,
    predicted_requests_per_second=30,
    expected_session_duration_minutes=30,
))

print('
Resource allocation estimate:')
print(recommend_resource_allocation(
    predicted_concurrent_users=150,
    predicted_requests_per_second=35.5,
    expected_session_duration_minutes=30,
))

print('
What-if combined simulation:')
print(run_what_if_simulation({**input_example}))
```
